#### Fit a Linear Regression Model

```python
import statsmodels.formula.api as smf
model1 = smf.ols("sunday~daily",data = data).fit()
model1.summary()
```

```python
#plot the scatter plot and overlay the fitted sttaight line using matplotlib
x = data["daily"].values
y = data["sunday"].values
plt.scatter(x, y, color = "m", marker = "o", s = 30)
b0 = 13.84
b1 = 1.33
y_hat = b0 + b1*x
plt.plot(x, y_hat, color = "g")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
sns.regplot(x="daily", y="sunday", data=data)
plt.xlim([0,1250])
plt.show()
```

#### Predict for new data points(test data)

```python
#predict sunday sales for 200 and 300 and 400 and 500 daily circulation
newdata=pd.Series([200,300,500])
data_pred=pd.DataFrame(newdata,columns=['daily'])
data_pred                                       
```

```python
model1.predict(data_pred)
```

```python
pred = model1.predict(data["daily"])
pred
                      
```

```python
data["Y_hat"] = pred
data
```

```python
data["residuals"]=data["sunday"]-data["Y_hat"]
data
```

```python
mse = np.mean((data["daily"]-data["Y_hat"])**2)
rmse = np.sqrt(mse)
print("MSE: ",mse)
print("RMSE: ",rmse)
```

```python
mae = np.mean(np.abs(data["daily"]-data["Y_hat"]))
mae
```
