```python
data = {
    'User ID': [1, 1, 2, 2, 3, 3, 4, 3, 7],
    'Movie Name': ['Inception', 'Titanic', 'Incepption', 'Avatar', \
                   'Titanic', 'Avatar', "Lion King", "Inter Stellar", "Bahubali"],
    'Rating': [9, 8, 7, 10, 9, 8, 7, 8, 9]
}
#Create DataFrame
df = pd.DataFrame(data)
print(df)
                   
```

```python
#create a sample table with movie ratings
data = {
    'User ID': [1, 1, 2, 2],
    'Movie Name': ['Apple', 'Banana', 'orange', 'Grapes'],
    'Rating': [9, 8, 7, 10]
}
#Create DataFrame
df = pd.DataFrame(data)
print(df)
```

```python
#pivot the table
pivot_table = df.pivot(index='User ID', columns='Movie Name', values='Rating')
#Dispaly the pivoted table
print(pivot_table)
```
